```r
library ("tidyverse")

data(mpg)
mpg
```

```r
mpg %>% # pipe operator "luego"
 add_row(manufacturer = "jeep",
         displ        = 2,
         year         = 2008,
         cyl          = 4, 
         trans        = "manual=m6",
         cty          = 15,
         hwy          = 24,
         fl           = "p",
         class        = "suv"
          )
```

```r
mpg  %>%

  mutate(

    cty = cty * 1.609 / 3.785,

    hwy = hwy * 1.609 / 3.785,

    manufacture = as_factor(manufacturer),

    model = as_factor(model),

  )
```

```r
mpg_editada <- mpg  %>% 
 mutate_if(
   .predicate = is.character,
   .funs      = as_factor
 )
mpg_editada
```

```r
mpg_editada2 = mpg %>%
  mutate_if(
  .predicate = is.character,
  .funs      = as_factor
  )  %>%
  mutate(trans = fct_lump_min(trans, 20,
                             other_level = "Otros"))
mpg_editada2
```

```r
mpg_editada2 %>%
  ggplot() +
  geom_point(mapping = aes(x = displ, y = hwy))
```

```r
mpg_editada2  %>%
ggplot() +
  geom_point(mapping = aes(x = displ,
                           y = hwy,
                           color = class))
```

```r
ggplot(data = mpg_editada2) +
geom_point(mapping = aes(x = displ,
                          y = hwy,
                          color = class,
                          shape = drv,
                          size = cyl),
           alpha = 0.7)
```
